# Import important lib and download LLM model ( Text generate )

```python
import torch 
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

model_name = "scb10x/llama-3-typhoon-v1.5x-8b-instruct"

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    device_map="auto",
    torch_dtype=torch.float16,
    trust_remote_code=True,
)

tokenizer = AutoTokenizer.from_pretrained(model_name)

# Explicitly set pad_token_id to eos_token_id
tokenizer.pad_token_id = tokenizer.eos_token_id

pipe = pipeline("text-generation", model=model, tokenizer=tokenizer)

generation_args = {
    "max_new_tokens": 512,
    "return_full_text": False,
    "pad_token_id": tokenizer.pad_token_id, 
}
```

# Function Called LLM

```python
def query(messages):

  output = pipe(messages, **generation_args)
  return output[0]['generated_text']
```

Chat modify function ( loop ) ( very high GPU and memory usage )

```python
def chat():

  conversation_history = [
      {"role": "system", "content" : "You are Dev God , you can plan for model business , you are very good at business ,you will give every single details about business that user ask"
      }]
  while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        break

    conversation_history.append({"role": "user", "content": user_input})

    try:
        response = query(conversation_history)
        print("Teacher: ", response)

        conversation_history.append({"role": "assistant", "content": response})

    except Exception as e:
        print(f"An error occurred: {e}, please try again.")
```

```python
chat()
```
